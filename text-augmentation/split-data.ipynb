```python
import pandas as pd
```

```python
DATASET_PATH = '../dataset/OBC_Cleaned.csv'

GUILTY_DATASET_PATH = '../dataset/split/guilty_split.csv'
NOT_GUILTY_DATASET_PATH = '../dataset/split/not_guilty_split_X.csv'
```

```python
df = pd.read_csv(DATASET_PATH)

df.head()
```

```python
df['Verdict'].value_counts()
```

```python
guilty_df = df[df['Verdict'] == 'guilty'].copy()
guilty_df = guilty_df.reset_index(drop=True)
df[['Trial_Text']] = df[['Trial_Text']].replace('\n', '', regex=True)

guilty_df.head()
```

```python
guilty_df.shape
```

```python
guilty_df.to_csv(GUILTY_DATASET_PATH, index=False)
```

```python
not_guilty_df = df[df['Verdict'] == 'notGuilty'].copy()
not_guilty_df = not_guilty_df.reset_index(drop=True)

not_guilty_df.head()
```

```python
not_guilty_df.shape
```

```python
CHUNK_SIZE = 1000
total_chunks = len(not_guilty_df) // CHUNK_SIZE + (1 if len(not_guilty_df) % CHUNK_SIZE != 0 else 0)
```

```python
for i in range(total_chunks):
    start_idx = i * CHUNK_SIZE 
    end_idx = start_idx + CHUNK_SIZE
    chunk = not_guilty_df[start_idx:end_idx]
    chunk.to_csv(NOT_GUILTY_DATASET_PATH.replace('X', str(i+1)), index=False)
```
